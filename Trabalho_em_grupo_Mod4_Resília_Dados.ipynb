```python
#4. Qual o preço médio da gasolina e do etanol por estado?
# Cálculo das médias de preço de gasolina e etanol por estado
media_por_estado_gasolina = df_combinedDados[df_combinedDados['Produto'] == 'GASOLINA'].groupby('Estado - Sigla')['Valor de Venda'].mean()
media_por_estado_etanol = df_combinedDados[df_combinedDados['Produto'] == 'ETANOL'].groupby('Estado - Sigla')['Valor de Venda'].mean()

# Gráficos de barras para preço médio por estado
plt.figure(figsize=(12, 6))
media_por_estado_gasolina.plot(kind='bar', color='blue', label='Gasolina')
media_por_estado_etanol.plot(kind='bar', color='orange', label='Etanol')
plt.xlabel('Estado')
plt.ylabel('Preço Médio')
plt.title('Preço Médio da Gasolina e Etanol por Estado')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```

```python
# 5. Qual o município que possui o menor preço para a gasolina e para o etanol?

# Identificação dos municípios com menor preço de gasolina e etanol
municipio_menor_preco_gasolina = df_combinedDados[df_combinedDados['Produto'] == 'GASOLINA'].sort_values('Valor de Venda').iloc[0]['Municipio']
municipio_menor_preco_etanol = df_combinedDados[df_combinedDados['Produto'] == 'ETANOL'].sort_values('Valor de Venda').iloc[0]['Municipio']

# Exibição dos resultados dos municípios com menor preço
print(f"Município com Menor Preço de Gasolina: {municipio_menor_preco_gasolina}")
print(f"Município com Menor Preço de Etanol: {municipio_menor_preco_etanol}")
```

```python
# 6. Qual o município que possui o maior preço para a gasolina e para o etanol?

# Identificação dos municípios com maior preço de gasolina e etanol
municipio_maior_preco_gasolina = df_combinedDados[df_combinedDados['Produto'] == 'GASOLINA'].sort_values('Valor de Venda', ascending=False).iloc[0]['Municipio']
municipio_maior_preco_etanol = df_combinedDados[df_combinedDados['Produto'] == 'ETANOL'].sort_values('Valor de Venda', ascending=False).iloc[0]['Municipio']

# Exibição dos resultados dos municípios com maior preço
print(f"Município com Maior Preço de Gasolina: {municipio_maior_preco_gasolina}")
print(f"Município com Maior Preço de Etanol: {municipio_maior_preco_etanol}")
```

```python
# 7. Qual a região que possui o maior valor médio da gasolina?

# Cálculo da média de preço de gasolina por região e seleção da região com maior preço médio
media_por_regiao_gasolina = df_combinedDados[df_combinedDados['Produto'] == 'GASOLINA'].groupby('Regiao - Sigla')['Valor de Venda'].mean()
regiao_maior_preco_gasolina = media_por_regiao_gasolina.idxmax()

# Exibição do resultado da região com maior preço médio
print(f"Região com Maior Preço Médio de Gasolina: {regiao_maior_preco_gasolina}")
```

```python
# 8. Qual a região que possui o menor valor médio do etanol?

# Cálculo da média de preço de etanol por região e seleção da região com menor preço médio
media_por_regiao_etanol = df_combinedDados[df_combinedDados['Produto'] == 'ETANOL'].groupby('Regiao - Sigla')['Valor de Venda'].mean()
regiao_menor_preco_etanol = media_por_regiao_etanol.idxmin()

# Exibição do resultado da região com menor preço médio
print(f"Região com Menor Preço Médio de Etanol: {regiao_menor_preco_etanol}")
```

```python
# 9- Há alguma correlação entre o valor do combustível (gasolina e etanol) e a região onde ele é vendido?

# Filtrar apenas os dados de gasolina e etanol
df_gasolina = df_combinedDados[df_combinedDados['Produto'] == 'GASOLINA']
df_etanol = df_combinedDados[df_combinedDados['Produto'] == 'ETANOL']

# Calcular a média dos preços por região
media_por_regiao_gasolina = df_gasolina.groupby('Regiao - Sigla')['Valor de Venda'].mean()
media_por_regiao_etanol = df_etanol.groupby('Regiao - Sigla')['Valor de Venda'].mean()

# Criar um DataFrame com as médias de preços por região
df_regiao = pd.DataFrame({'Gasolina': media_por_regiao_gasolina, 'Etanol': media_por_regiao_etanol})

# Calcular a correlação entre os preços de gasolina e etanol por região
correlation_regiao = df_regiao['Gasolina'].corr(df_regiao['Etanol'])

# Criar um gráfico de dispersão para visualizar a correlação
sns.scatterplot(data=df_regiao, x='Gasolina', y='Etanol')
plt.xlabel('Preço Médio de Gasolina')
plt.ylabel('Preço Médio de Etanol')
plt.title('Correlação entre Preços de Gasolina e Etanol por Região')
plt.annotate(f'Correlação: {correlation_regiao:.2f}', xy=(0.15, 0.85), xycoords='axes fraction')
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
# 10 - Há alguma correlação entre o valor do combustível (gasolina e etanol) e a bandeira que vende ele?

# Calcular a média dos preços por bandeira
media_por_bandeira_gasolina = df_gasolina.groupby('Bandeira')['Valor de Venda'].mean()
media_por_bandeira_etanol = df_etanol.groupby('Bandeira')['Valor de Venda'].mean()

# Criar um DataFrame com as médias de preços por bandeira
df_bandeira = pd.DataFrame({'Gasolina': media_por_bandeira_gasolina, 'Etanol': media_por_bandeira_etanol})

# Calcular a correlação entre os preços de gasolina e etanol por bandeira
correlation_bandeira = df_bandeira['Gasolina'].corr(df_bandeira['Etanol'])

# Criar um gráfico de dispersão para visualizar a correlação
sns.scatterplot(data=df_bandeira, x='Gasolina', y='Etanol')
plt.xlabel('Preço Médio de Gasolina')
plt.ylabel('Preço Médio de Etanol')
plt.title('Correlação entre Preços de Gasolina e Etanol por Bandeira')
plt.annotate(f'Correlação: {correlation_bandeira:.2f}', xy=(0.15, 0.85), xycoords='axes fraction')
plt.grid(True)
plt.tight_layout()

plt.show()
```
