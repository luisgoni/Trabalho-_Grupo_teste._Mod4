## Copiando os dados para a Análise

```python
#Copiando o dataFrame original para analisar
df_maio = Dados_mes_maio.copy()
df_junho = Dados_mes_junho.copy()
```

## Convertendo os dados de Valor de Venda para numeros floats e com separaçao por ponto

```python
df_maio['Valor de Venda'] = df_maio['Valor de Venda'].str.replace(',','.', regex=True).astype(float)
df_junho['Valor de Venda'] = df_junho['Valor de Venda'].str.replace(',','.', regex=True).astype(float)
```

- Observando valores outlier nos preços dos combustiveis do mes de maio


```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_maio, x='Valor de Venda', y='Produto')
plt.title('Distribuição dos Preços da Gasolina por Produto')
plt.xlabel('Valor de Venda Mês de Maio')
plt.ylabel('Produto')
plt.show()
```

## Separando os produtos existentes d mês de Maio;
- Produto Gasolina Aditivada

```python
Produto_Gasolina_AD05 = df_maio[df_maio['Produto']=='GASOLINA ADITIVADA']
Produto_Gasolina_AD05.head()
```

- Observando os 5 primeiros dados do preço so da Gasolina aditivada.

```python
Produto_Gasolina_AD05[['Produto','Valor de Venda']].head()
```

- Preço Variado da  Gasolina Aditivada

```python
Preço_max_GasAD = Produto_Gasolina_AD05['Valor de Venda'].max()
print(f'O Preço maximo da gasolina aditivada foi de {Preço_max_GasAD}')
Preço_min_GasAD = Produto_Gasolina_AD05['Valor de Venda'].min()
print(f'O Valor minimo vendido pela gasolina aditivada e de {Preço_min_GasAD}')
Preço_medio_GasAD = Produto_Gasolina_AD05['Valor de Venda'].mean()
print('O preço medio vendido sobre a gasolina aditivada e de {:.2f}'.format(Preço_medio_GasAD))
```

- Observando a analise atraves de grafico

```python
plt.figure(figsize=(15, 10))
sns.lineplot(data=Produto_Gasolina_AD05, x='Data da Coleta', y='Valor de Venda')
plt.title('Variação dos Preços da  Gasolina Aditivada ao longo do Mês de Maio')
plt.xlabel('Data da Coleta')
plt.ylabel('Valor de Venda')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
